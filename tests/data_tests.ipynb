```python
import matplotlib.pyplot as plt
import random

from data_loading import BavarianBuildingDataset
```

```python
dataset = BavarianBuildingDataset(image_dir='../data/bbd250-image', mask_dir='../data/bbd250-umring')
```

```python
# Display six randomly selected image-mask pairs in four columns
indices = random.sample(range(len(dataset)), 6)

fig, axes = plt.subplots(3, 4, figsize=(8, 6))

for k, index in enumerate(indices):
    i, j = divmod(k, 2)
    image, mask = dataset.get_image_mask_pair(index)
    
    axes[i, j*2].imshow(image)
    axes[i, j*2].set_title('Image')
    axes[i, j*2].axis('off')

    axes[i, j*2 + 1].imshow(mask)
    axes[i, j*2 + 1].set_title('Mask')
    axes[i, j*2 + 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Compare the sizes of the training, test, and validation sets
train_indices, test_indices, val_indices = dataset.train_test_val_split()

print(f'Training set size: {len(train_indices)}')
print(f'Test set size: {len(test_indices)}')
print(f'Validation set size: {len(val_indices)}')
```
