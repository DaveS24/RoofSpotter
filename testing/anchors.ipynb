```python
import numpy as np

np.random.seed(0)

anchors = np.array([[0.5, 0.5],
                    [0.5, 0.75],
                    [0.5, 1.],
                    [0.5, 1.5],
                    [0.5, 2.],
                    [0.5, 2.25],
                    [0.5, 3.],
                    [0.5, 4.],
                    [1., 0.5],
                    [1., 0.75],
                    [1., 1.],
                    [1., 1.5],
                    [1., 2.],
                    [1., 2.25],
                    [1., 3.],
                    [1., 4.]])

offsets = np.random.uniform(low=-2, high=2, size=(64))
offsets[2::4] = np.abs(offsets[2::4])
offsets[3::4] = np.abs(offsets[3::4])

print(anchors)
print(offsets)
```

```python
print(anchors.shape)
print(offsets.shape)
```

```python
for x in range(2):
    for y in range(2):
        anchor_w = anchors[:, 0]
        anchor_h = anchors[:, 1]

        dx = offsets[::4]
        dy = offsets[1::4]
        dw = offsets[2::4]
        dh = offsets[3::4]

        ctr_x, ctr_y = x + dx, y + dy # Center of the proposal
        w, h = anchor_w * np.exp(dw), anchor_h * np.exp(dh) # Width and height of the proposal

        print(ctr_x)
        print('-----')
        print(ctr_y)
        print('-----')
        print(w)
        print('-----')
        print(h)
        print('######')

        x1, y1 = ctr_x - 0.5 * w, ctr_y - 0.5 * h
        x2, y2 = ctr_x + 0.5 * w, ctr_y + 0.5 * h

        print(x1)
        print('-----')
        print(y1)
        print('-----')
        print(x2)
        print('-----')
        print(y2)

        assert all(x1 < x2), "x1 is not smaller than x2"
        assert all(y1 < y2), "y1 is not smaller than y2"
```
